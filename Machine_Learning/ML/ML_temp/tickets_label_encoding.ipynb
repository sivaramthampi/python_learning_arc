```python
import pandas as pd
import numpy as np 
df = pd.read_csv('ticket_08_2017 - ticket_08_2017.csv')
df
```

```python
df1 = df.iloc[:,6:13]
df1
```

```python
df1.columns
```

```python
df2 = df1.groupby(['sync_tkt_route_name', 'sync_tkt_schedule_no', 'sync_tkt_trip_no',
                   'sync_tkt_start_stage_name', 'sync_tkt_end_stage_name'],as_index=False).agg({'sync_tkt_full_count':'sum'})
df2
```

```python
x = df2.iloc[:,[0,1,2,3,4]].values
y = df2.iloc[:,[-1]]
x
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression 
from sklearn.model_selection import train_test_split
```

```python
label=LabelEncoder()
x[:,0] = label.fit_transform(x[:,0])
x[:,3] = label.fit_transform(x[:,3])
x[:,4] = label.fit_transform(x[:,4])
```

```python
x
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y)
model = LinearRegression()
model.fit(xtrain,ytrain)
y_pred = model.predict(xtest)
```

```python
y_pred
```

```python
model.score(xtest,ytest)
```
